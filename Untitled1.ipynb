```python
import pandas as pd
from pydantic import BaseModel, Field, ValidationError
from typing import List

class ResponseChecks(BaseModel):
    data: List[str] = Field(..., min_items=1)

def extract_keywords(keywords: pd.DataFrame, listings: pd.DataFrame, llm):
    # Convert keywords DataFrame to a comma-separated string
    keywords_str = ', '.join(keywords.apply(lambda row: ', '.join(row.dropna().astype(str)), axis=1))
    
    # Convert listings DataFrame to a comma-separated string
    listings_str = ', '.join(listings.apply(lambda row: ', '.join(row.dropna().astype(str)), axis=1))
    
    # Create the prompt
    prompt = f"""Using the following keywords data as a reference:

{keywords_str}

Please extract relevant keywords from the following listings data. 
For each listing, provide the listing text followed by a hyphen and then the extracted keywords, separated by commas. 
For example: 'Listing text - keyword1, keyword2, keyword3'

Listings data:
{listings_str}"""

    # Invoke the language model
    response = llm.invoke(prompt)
    
    # Process the response
    response_lines = response.split('\n')
    
    # Keep only the lines in between blank lines (removing any explanation lines)
    blank_indexes = [index for index in range(len(response_lines)) if response_lines[index] == '']
    if len(blank_indexes) == 1:
        response_lines = response_lines[(blank_indexes[0] + 1):]
    elif len(blank_indexes) > 1:
        response_lines = response_lines[(blank_indexes[0] + 1) : blank_indexes[1]]
    
    # Print response and validate if it is in the correct format
    print(response_lines)
    try:
        ResponseChecks(data=response_lines)
    except ValidationError as e:
        print(f"Validation error: {e}")
        return None

    # Put in dataframe
    keywords_df = pd.DataFrame({'Listing vs keywords': response_lines})
    keywords_df[['Listing', 'Keywords']] = keywords_df['Listing vs keywords'].str.split(' - ', expand=True)
    
    return keywords_df
```
